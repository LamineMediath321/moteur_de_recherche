### Les pré traitements 

```python
# On supprime les articles qui n'ont pas de contenu article et de resume
data.dropna(subset=["contenuArticle"], inplace = True)
data.dropna(subset=["resumeArticle"], inplace = True)
```

## Les mots vides

```python
import spacy
nlp = spacy.load("fr_core_news_sm")
# Les mots vides de notre jeu de données
french_stopwords = spacy.lang.fr.stop_words.STOP_WORDS
```

```python
# Ajouter quelques mots vides
french_stopwords.add('mal')
french_stopwords.add('ensuite')
french_stopwords.add('veut')
french_stopwords.add('jamais')
```

```python
# Définition de Fonction doc_without_stop_word pour le pre-traitement des textes
def doc_without_stop_word(text):
    doc = nlp(text)
    filtered = [str(token.text).lower() for token in doc if token.is_stop==False and token.text.isalpha()==True]
    filtered_text = ' '.join(filtered)
    return filtered_text
```

```python
# Le contenu sans mots vides
data['text_filtre'] = data['resumeArticle'].apply(lambda x: doc_without_stop_word(x))
data
```

### Lemmatization

```python
## Une foction qui lemmatise un text
def lemma(text):
    doc = nlp(text)
    filtered_text = ""
    for token in doc:
        if filtered_text=="":
            filtered_text = filtered_text+token.lemma_
        else:
            filtered_text = filtered_text+' '+token.lemma_
    return filtered_text
```

```python
# Le contenu sans mots vides
data['text_filtre'] = data['text_filtre'].apply(lambda x: lemma(x))
data.to_csv('articles/article_traiement.csv')
```

### On récupère le csv genere pour eviter de refaire le meme travail 

```python
## data apres traitement
data = pd.read_csv('articles/article_traiement.csv')
##La taille du document📄 
N_doc = len(data)
```

### La requête 

```python
requete = "2 éme EDITION DU FESTIVAL DAKAR KIZOMBA"
```

##### Les prétraitements concernant la requête 

```python
# Liste des termes à retenir de la requête
requete_filtre=[]
# Traiter la requête
doc = nlp(requete)
requete_filtre = doc_without_stop_word(doc)
requete_filtre = lemma(requete_filtre)
requete_filtre
```

```python
## On ajoute la requete à la fin du dataframe
data=data.append({'identifiantArticle' : 'la requete','text_filtre' : requete_filtre}, ignore_index=True)
data
```

### TF-IDF

### Représentation des données 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(sublinear_tf=True,min_df=5,max_df=0.95)
```

```python
tfidf_dtm=vectorizer.fit_transform(data['text_filtre'])
tfidf_dtm.shape
```

```python
termsVect = vectorizer.get_feature_names()
tdidf_df = pd.DataFrame(tfidf_dtm.toarray(), columns=termsVect)
tdidf_df
```

## Calcul de similarité entre notre corpus et une requête

```python
from scipy.spatial import distance
```

```python
##Recuperer la requete
N = len(data)-1
```

## vecteur de la requete

```python

##Le dernier element du dataFrame
request = tdidf_df.iloc[N,:].values
request
```

## Similarité cosinus
### On calcul les documents les plus similaire par rapport à la requete

```python
def similaire(tab_vector,requete,taille,topN=50):
    my_df = []
    taille = taille-1
    for i in range(taille):
        # vecteur de chaque texte
        doc_vect = tab_vector.iloc[i,:].values
        # distance cosinus entre les deux vecteurs
        cos_dist = distance.cosine(doc_vect.tolist(), request.tolist())
        sim_cos = 1 - cos_dist
        if sim_cos<1:
            d = {
                'identifiantArticle': data.iloc[i,1],
                'titreArticle': data.iloc[i,6],
                'similarite': sim_cos,
                'contenuArticle': data.iloc[i,7]
            }
            my_df.append(d)
    ##On trie
    my_similaire = pd.DataFrame(my_df)
    ##Trier du plus pertinent au moins pertinent 
    my_similaire = my_similaire.sort_values(by = 'similarite', ascending = False)
    ##On retourne les topN les plus pertinents 
    return my_similaire.iloc[0:topN]
```

### Les documents les plus pertinents 

```python
plus_pertinent = similaire(tdidf_df,request,N_doc,10)
plus_pertinent
```

#### Les documents pertinents en json

```python
doc_json = plus_pertinent.to_json(orient='records')
doc_json
```
